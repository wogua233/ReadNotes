Linux——编译应该使用g++编译器自动工作

Python 3.7包含以下包:Numpy, matplotlib和PyQt5(用于GUI)
GCC 9.3或更高版本，或者clang 10.0或更高版本。

DFT的输入：为了效率更高，将最上方的原子序号改为1

在.inp中 FORCE_EVAL → DFT → PRINT 加入:

      &MO ON
        EIGVECS
        CARTESIAN
        FILENAME cartesian-mos
        &EACH
          QS_SCF 0
        &END EACH
      &END MO

对 CP2K 和 FHI-AIMS 自旋极化的计算也可用. 可以调整 'alpha' = 'up' spin, 'beta' = 'down' spin or 'both' spins contributing to your current.

模拟的种类：

didv - single dI/dV (conductance) calculation at the energy (relative to the Fermi Level) given by V .

USE 'V-scan' instead of 'STM'：V-scan使您将在一次运行中获得Vmin和Vmax之间的所有内容，调整cut_min and cut_max如-2~+2

For molecules: New feature 'states'：你不需要自己指定那些特征能量;脚本会自己找到它们，在计算dI/dV时仍然考虑了所有的状态。如果您只想使用一个状态，请选择非常低的eta(例如0.0000001)

注意：

脚本不能使用多个k点(它只能读取伽玛k点数据)，因此使用的supercell必须足够大，这样伽玛点才能正确描述样本的电子状态

脚本不会将扫描点折叠到原始单元中，因此你可以(有时)找到几何体的“末端”，在那里发生“边界效应”。如果电子结构正确(!)，这些效应不能渗透到原来的超级电池内部，但它们会影响到几何末端附近几埃

样品的轨道与DFT计算中使用的LCAO基集没有直接联系。尽管你可以对任何元素使用很大的基集，但PPSTM代码只使用价轨道。
因此样品轨道可以限于sp

由于样品的d轨道比sp轨道衰变到真空的速度更快，因此它们使用0.2的常数重新标度来“阻尼”

该因子来源于FHI-AIMS的s轨道和d轨道的比较。重新缩放并不理想，但是不同的DFT代码对不同的元素使用不同的衰减，因此我们在该代码的哲学中没有发现其他简单的一般规则

您还可以通过比较计算中单个高度的tip_orb = s的dIdV扫描来研究sp和spd之间的差异

Cut_atoms选项：可用于(大大)加快计算速度(特别是对于厚板和高PBC选项)。在大多数情况下，您可以切断所有最上面的(重要的)原子3 Å以下的原子

所谓重要，我们指的是任何具有相当多状态的东西。也就是说，在平面有机结构上，氢的贡献可能比C原子要小得多，那么C原子就算数了

如果原子之间的电负性差异对电流的贡献很大，总是检查这些原子上面的哈特里电位，如果顶层原子上方的电势发生了较大变化，请尝试估计lower_atoms和_lower_coefs_(对高电负性原子重新缩放)例如，对于TOAT分子羰基上的O，氧原子的下限系数设为0.5( PRB 95, 045407 (2017) ).

输出可用于不同的高度，电压，扫描技术(STM或dIdV)和尖端原子轨道:

tip原子轨道有:s, px和py, pz, dz2, dxz和dyz。最后两个选项只适用于样品的sp轨道，而前三个选项同时适用于样品的sp和spd轨道。

s的尖端原子轨道几乎与Tersoff-Hamann理论(PRL 50, 1998(1983))一样——它超越了这个理论，因为它考虑了干涉效应(as e.g. in ACS Nano 9 (9), pp 9180–9187 (2015) )

ppstm并不直接从dft中计算针尖和样品的电流，tip的轨道是依据经验单独构造的：可选用不同成分的组合

All the 'fixed' scans are constant-height scans，fixed代表恒高扫描。应当设置在最高原子的4~6埃左右。要运行单高度扫描，使用[zmin, zmax>zmax+dz, dz]。

对于恒流，包没有提供直接绘制的方法，可以利用ppstm绘制最少2A，间隔0.1~0.2A的数据，在VESTA中绘制。由于I或dIdV ~ Exp(-beta z)，您可以尝试以对数尺度绘制图像(未在PPSTM_simple.py中编程，您可以使用WSXM)来模拟恒流扫描

具体参数：

codes:Fireball, FHI-AIMS, CP2K and GPAW

sample orbital:sp或是spd

spin:闭壳层none，自旋极化up, down or both

pbc：超胞

lvs:二维点阵向量
