# Can Orbitals Really Be Observed in Scanning Tunneling Microscopy Experiments?

所有可观察到的原子或分子性质都是由概率和相应的量子力学算符决定的，而不是由波函数本身决定的。波函数，甚至精确波函数，都是不可观测的。

为了得到波函数的近似，通常采用单电子函数（轨道）的概念，这是一种平均场近似的思想，在HF和DFT方法中都有体现

此外，由于波函数的任意幺正变换不会改变其振幅（概率密度），因此也不会改变任何可观察到的性质，因此用于构建波函数的MO基不是唯一的。例如，两组完全不同的正则化和局域化MOs可以形成基来构建具有相同密度的波函数，它们代表系统的相同电子状态。因此，MOs不是唯一的，也是不可观测的

报导经常出现‘观测到分子的前线轨道’，这种说法是不正确的：(i)系统的任何性质只能由总密度完全表示；（ii）轨道是简单的数学构造，用来建立（近似的）波函数，然后是密度；（iii）因此，轨道不能直接与可观察到的分子性质相关联；（iv）轨道是非唯一的，因为在给定的子空间（例如，双占位空间）内（HF或KS）轨道之间的任何幺正变换的能量是不变的。

# Electronic and Chemical Properties of Donor, Acceptor Centers in Graphene

由于量子干涉效应，可以在STM通道中实现硼和氮取代缺陷的化学识别

B和N元素分别是在石墨烯中常用的p-和n-型掺杂

通过1200℃高温处理SiC（B流掺杂）,得到掺B石墨烯。进一步在此基础上使用离子注入方法掺杂N

N中心有一凹陷，而B中心高亮，这归因于N中心的相消干涉

由等高扫描构建的3DSTM

干涉效应导致N中心的电流-距离曲线远离了隧穿的指数衰减形状，参与隧穿的前线轨道表明，在N的附近有反键特征，在中心N原子和C之间有波函数节点。考虑了相消干涉的STM模拟是通过实际考虑针尖来进行计算的（FIREBALL格林函数方法-输运）

# The Kondo resonance line shape in scanning tunnelling spectroscopy: instrumental aspects

由STS谱图中kondo的线形可以得到：1、Kondo系统的内在性质（受温度和外磁场调控）；2、隧穿的矩阵元效应，可由不同隧穿通道的干涉来解释；3、仪器展宽。

在STS中的近藤共振峰在升温和磁场作用下的分裂和扩散（有可能分别发生），有时被用于将kondo和其他STS零能特征区分开来。

首先，STS的kondo峰应符合fano线形；

V-样品电压，$\epsilon$-归一化能量，$E_k$是共振的能量，$\Gamma$是半高宽half-width at half-maximum (HWHM)

不对称因子q改变了共振的线形，从不对称特征上的洛伦兹峰（q→∞）到dip（q→0）

1986年Frota提出一个不同的线形，Frota线形与数值重整化群（NRG）计算得出的近藤共振非常吻合 

尽管与NRG计算预测的对数衰减相比，Frota线的尾部随能量的平方根衰减，但Frota线的形状与NRG计算相对较好地匹配，并且在较宽的能量范围内

$\Gamma$也是半高宽，$\phi$是一个因子决定了线形

近藤效应的特征能量尺度通常由近藤温度来表示，其与STS的光谱宽度直接相关，这就要求将展宽的来源考虑清楚

使用视黄酸在Au(111)表面做模型研究：kondo在这个体系中干净且可以调控宽度
针尖操纵后形成两个自旋中心

使用低通滤波减小了kondo峰的展宽

电压调制带来的噪声对kondo谱的影响是相当显著的，且低通滤波的高质量谱更适合fano拟合，而未经低通滤波的带噪声的谱更适合frota拟合

对于另一个自旋中心，此处本征的kondo就有较大展宽，所以仪器展宽不那么明显，两个线性都更适合Frota你拟合

采集STS谱通常需要用到锁相放大器来提高信噪比，这项技术需要对样品加上一个额外的正弦调制电压$V_m$

越大的调制电压对应越大的展宽

且不同的拟合函数对应不同的半高宽，较小的调制Frota更合适，较大的调制Fano更合适

较大的样品电压和较小的调制电压有利于Frota,过度的调制幅度和有限的拟合范围会误导Fano线形状

另外：样品片偏压本身的噪声和针尖温度都有可能造成展宽
