各种相变的现象和原因错综复杂，而相变临界点附近的物理量的奇异性十分相似，长期以来，人们用平均场理论来描述连续相变，该理论的出发点是用一个平均了的场，即内场，来代替其他粒子对某个特定粒子的作用，从而把多体问题简化为单体问题。

埃伦费斯特对相变的分类方式是热力学势和其导数的连续性：凡是热力学势本身连续，而第一阶导数不连续的状态突变，称
一阶类相变。第一阶导数不连续，表示相变伴随着明显的体积
化和热量的吸放（潜热）。普通的气液相变和在外磁场中的超转
变，都是一类相变的实例。

热力学势和它的第一阶导数连续变化，而第二阶导数不连续的
情形，称为第二类相变。这时没有体积变化和潜热，但比热、压
率、磁化率等物理量随温度的变化曲线上出现跃变或无穷的尖。
超流（ λ 点）和没有外磁场的超导转变、气液临界点，以及量磁
相变，属于二类相变

习惯上，把二阶和更高阶的相变称为连续相变和临界现象

序参量：通常低温相对成型低有序度高，高文相对程度高有序度低，序参量在Tc以上为0，Tc以下不为0。连续相变指序参量在Tc连续的减小到零，对某些一阶相变，序参量在Tc阶跃到非0值一阶相变和连续相变

对于超导和超流转变，序参量不在明显，是一名为‘宏观波函数’或‘能隙’的负数，有模量和相位角，在正常态，模量为0，绕原点转动对称（U(1)规范对称），在超导/超流态，模量不为零，因此超导超流的序参量有两个内在自由度（模量和相位）。因此可见，序参量可以有很不相同的结构，可能是标量，也有可能是多维的变量，但是在临界点上绝对值趋于0是一样的。序参量可以和一定的外场耦合，称为‘对偶场’。（平均磁矩是自由能对磁场的导数）

对称的破缺和恢复:对称性质的突然降低，称之为对称的破缺，对称性质的有或无只有突变没有渐变，但是对称性的突变可以使由某个渐变引起的，物理参数的无穷小变化引起对称
的破缺，这是连续相变的本质

从高温高对称相到低温低对称相的转变必须通过某种破坏对称的运动来实现，如果各种运动模式同样地激发，不会破坏对称性

从高温相逼近临界点，各个运动模式都可能软化发生凝聚，并破坏对成型；从低温相逼近临界点，各个运动模式软化并恢复对称。软化指的是激发能趋近于0

能像晶核那样从旧相中逐渐地长大。在各向同性的铁磁体中，当
度远高于居里点时，相邻两点间磁矩的方向关联很弱。温度降低，
磁相互作用逐渐和热运动抗衡。如果盯住一个磁矩观察，则对邻
近的磁矩有越来越大的影响，要求它们转到与自己平行的向。但
是，这种考虑适用于体系中任何一个磁矩，因此平均磁始终等于
零，只是影响或关联的半径逐渐增大。

_卡丹诺夫提出普适类：_ 各种物理体系可以分成若干个普适类，每个普适类的临界
特性完全一样。区分普适类的最重要标志是空间维数 D，其次是
部自由度的数目或者说序参量的个数 n。如果是短程力，临界特与
作用力的性质无关；对于长程作用，还要考虑作用力随距离减的
快慢。晶体的对称和微观磁矩的大小等，都是无关紧要的因素

_重整化群：_ 几何相变中，是金属球的概率是P，一维线中，划分元胞，每个元胞b个格点，元胞导通的概率记为Rb(P)，显然只要P!=1，那么一定Rb一定会在迭代中趋于0

将Rb这些重整化的整体称为重整化群，群乘法的规则可写成先将b个格点归为小元胞，再将b个小元胞归为大元胞，等价于将b^2个格点归为大元胞。但是归并没有逆操作，因此重整化群严格上是一个半群

几何相变中定义的关联长度是连通簇的平均长度，关联长度在重整化变换下变短b倍，由于临界点Pc关联尺寸无限大，所以关联尺寸不再因重整变换而变化，到达不变点，但是不变点不代表到达临界点。例如此例中有两个不动点，P\*1=0和P\*2=1,由于b>1,P\*2=1是不稳定不动点，重整变换倾向于远离该不动点。在相变理论中，如果某个量的值随着重整变换变大，那么就称之为‘有关参数’，|P-Pc|就是有关参数

关联长度与将重整变换在不动点附近展开lambda是Rb在Pc处的一阶导数由此可求出临界指数v

由此，重整化群的计算是包括三步：1、找到恰当的重整化变换；2、找到恰当的不动点，找到跟临界有关的不动点和有关参数；3、分析在这个不动点附近的变换性质

在相变点附近，由进程相互作用主导的体系也可能发生长程关联，由于关联尺寸趋于无穷，必须同时考虑体系所有粒子的影响，这就是平均场理论失败的原因。
