**Lieb–Mattis 理论**：传统的Lieb–Mattis 理论说在仅考虑最近邻hop的时候，一维Hubbard模型不可能出现铁磁性。低自旋基态总是低于高自旋基态。

Ghosh证明了一维或二维的Hubbard模型在任何有限温度下都不会表现出与磁长程序相关的对称性破缺.Uhrig同样排除了一般平面磁有序的可能性。用同样的方法，也可以证明超导U(1)对称性破缺的不存在。

**半满系统（U>>t时）**：一维系统，hop作为微扰加入U系统，能够交换的自旋构型（即反铁磁）能量降低，成为反铁磁链

**Lieb 理论**：二分型：对于一个hubbard lattice可以分为两套亚晶格，hop强度只有在两套亚晶格之间不为0。lieb理论表述，一个具有二分性t!=0且U>0的系统，基态除了自旋简并没有其他简并，且基态自旋=两个亚晶格site数目之差/2

lieb定理对于所有U>0的hubbard模型都生效，但是不同大小的U可能带来截然不同的物理性质

**lieb 铁磁性**：CuO晶格白格点数是黑格点的两倍，二者构成一个bipartite

键处形成hop-t，每处都有U>0，lieb定理表明该系统有总自旋S=L^2/2

有限温度下磁化率的Kubo-Kishi界：二分型hubbard模型中，在有限温度下的半填充模型中，任何波矢量q的电荷和on-site配对磁化率都是有限的。这意味着该模型不表现出任何CDW有序或超导有序

**自旋偏离半填充**：对于一个3 sites 2电子系统，一个电子向上一个电子向下假设两种t，当U趋向无穷，公有以下几种态可选：（不包含on site）

（t=t'/2）当U足够大，会出现铁磁基态

**Nagaoka’s 铁磁**：具有无限大U且总电子数比格点数少1的（恰好一个空穴），不同格点就有hop，且满足连通性（任意电子构型出发，通过t可以到达所有构型） ，证明基态是铁磁排列，这代表强相关的hubbard模型能够带来铁磁态

**Mielke’s 铁磁flat-band 铁磁**：在一kagome lattice中，相邻格点有hop,U>0，总电子数少于$\Gamma$/3。总哈密顿的基态同时是Hhop和Hint的基态，mielke的平带铁磁理论表明，具有高度简并单电子基态的Hubbard模型呈现饱和铁磁性的充分必要条件，有研究表明，加入一些扰动后，仍保持铁磁性

**非奇异哈伯德模型中的铁磁性**：保证、电子数四分之一填充且t/s 和 U/s足够大，那么就有饱和铁磁的基态
