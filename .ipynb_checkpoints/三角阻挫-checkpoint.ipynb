# Geometrical frustration

几何受挫磁体的研究关注的是当晶格几何结构抑制简单、有序、低温自旋构型的形成时会发生什么。通常，几何挫折会产生基态的简并流形，而不是单一的稳定基态构型，从而导致液体和冰的磁性类似物

两种可能的原理：1、约束不足，弱连接的晶格；2、涌现，新自由度的动态生成。

常见的自选阻搓晶格，三角晶格、kagomi晶格、烧绿石晶格

对于Ising spin 简并的基态数是可数的，且可以定义玻尔兹曼剩余熵，这是因为低温下，只有基态可以存在，总的自旋旋转熵减去从高温到低温降低的熵值就是剩余熵

自旋冰：对于常压下的六方冰，氧原子位置固定，氢原子可以靠近其相邻氧原子的其中一个，代表了一种Ising的模式而基态所要满足的约束条件是，一个氧原子必须有两个靠近的氢和两个较远的氢作为邻居

剩余熵的存在是由于存在大量的基态，这个系统如何能在它们之间波动呢？一种极端的可能性发生在铁磁体中，其中所有的自旋在基态中都指向同一方向。为了达到一个不同的基态，系统的所有旋转都需要翻转并一起重新定向。另一个极端发生在一个完全不相互作用的顺磁体中，每个自旋都可以单独翻转，因为能量并不依赖于它的方向
